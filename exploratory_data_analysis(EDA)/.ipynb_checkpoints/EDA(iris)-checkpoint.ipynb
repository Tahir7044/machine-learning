```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from statsmodels import robust
```

```python
data_set=pd.read_csv("iris/Iris.csv")
```

```python
#drop a column if axis=1 then column-wise else if axis=1 row-wise,and use the keyword inplace to make effect on data_set

data_set.drop(['Id'], axis=1,inplace=True)
```

```python
#to rename the columns of dataset 

data_set=data_set.rename(columns={"SepalLengthCm": "sepal_length", "SepalWidthCm": "sepal_width","PetalLengthCm": "petal_length", "PetalWidthCm": "petal_width"})
```

```python
#2-D scatter plots
#always undestand the axis: labels and scale.

data_set.plot(kind='scatter',x='sepal_length',y='sepal_width');
plt.show()
#cannot make any sense out of it.
```

```python
#lets see in color format
sns.set_style('whitegrid');
sns.FacetGrid(data_set,hue='Species',size=5).map(plt.scatter,'sepal_length','sepal_width').add_legend();
plt.show()
```

```python
#pair plot
plt.close();
sns.set_style('whitegrid');
sns.pairplot(data_set,hue='Species',height=7);
plt.show()
```

```python
iris_setosa=data_set.loc[data_set['Species']=='Iris-setosa']
iris_virginica=data_set.loc[data_set['Species']=='Iris-virginica']
iris_versicolor=data_set.loc[data_set['Species']=='Iris-versicolor']
#1-D scatter plot
plt.plot(iris_setosa['petal_length'],np.zeros_like(iris_setosa['petal_length']));
plt.plot(iris_virginica['petal_length'],np.zeros_like(iris_virginica['petal_length']));
plt.plot(iris_versicolor['petal_length'],np.zeros_like(iris_versicolor['petal_length']));
plt.show()
```

```python
print(iris_virginica.shape)
print(iris_setosa.shape)
print(iris_versicolor.shape)
```

```python
#distribution plot and density plot and univarent(only one feature consider at a time) analysis
#x-axis=petal length and y=axis=count the number of points in interval of x-axis
sns.FacetGrid(data_set,hue='Species',size=5).map(sns.distplot,'petal_length').add_legend();
sns.FacetGrid(data_set,hue='Species',size=5).map(sns.distplot,'petal_width').add_legend();
sns.FacetGrid(data_set,hue='Species',size=5).map(sns.distplot,'sepal_length').add_legend();
sns.FacetGrid(data_set,hue='Species',size=5).map(sns.distplot,'sepal_width').add_legend();
plt.show()
```

```python
#plot CDF of petal_length
counts_setosa,bin_edges_setosa=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
counts_virginica,bin_edges_virginica=np.histogram(iris_virginica['petal_length'],bins=10,density=True)
counts_versicolor,bin_edges_versicolor=np.histogram(iris_versicolor['petal_length'],bins=10,density=True)
pdf_setosa=counts_setosa/(sum(counts_setosa))
pdf_virginica=counts_virginica/(sum(counts_virginica))
pdf_versicolor=counts_versicolor/(sum(counts_versicolor))
#compute CDF
cdf_setosa=np.cumsum(pdf_setosa)
cdf_virginica=np.cumsum(pdf_virginica)
cdf_versicolor=np.cumsum(pdf_versicolor)
plt.plot(bin_edges_setosa[1:],pdf_setosa)
plt.plot(bin_edges_setosa[1:],cdf_setosa)
plt.plot(bin_edges_virginica[1:],pdf_virginica)
plt.plot(bin_edges_virginica[1:],cdf_virginica)
plt.plot(bin_edges_versicolor[1:],pdf_versicolor)
plt.plot(bin_edges_versicolor[1:],cdf_versicolor)
plt.show()
```

```python
#mean varience and std-deviation
print("Mean")
print(np.mean(iris_setosa['petal_length']))
print(np.mean(iris_versicolor['petal_length']))
print(np.mean(iris_virginica['petal_length']))
print("Std-dev")
print(np.std(iris_setosa['petal_length']))
print(np.std(iris_versicolor['petal_length']))
print(np.std(iris_virginica['petal_length']))
print("Median")
print(np.median(iris_setosa['petal_length']))
print(np.median(iris_versicolor['petal_length']))
print(np.median(iris_virginica['petal_length']))
```

```python
print("Quantiles")
print(np.percentile(iris_setosa['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_virginica['petal_length'],np.arange(0,100,25)))
```

```python
print("90th Quantiles")
print(np.percentile(iris_setosa['petal_length'],90))
print(np.percentile(iris_versicolor['petal_length'],90))
print(np.percentile(iris_virginica['petal_length'],90))
```

```python
print("Median Absolute Deviation")
print(robust.mad(iris_setosa['petal_length']))
print(robust.mad(iris_versicolor['petal_length']))
print(robust.mad(iris_virginica['petal_length']))
```
